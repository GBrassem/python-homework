```python
""" PyBank CSV Analysis
The purpose of this program is to analyze the budget data of a company over seven years
and determine net and average gains/losses, as well as the months of greatest profit and loss.
"""

from cgitb import text
from pathlib import Path
import csv

#Set the file path for the CSV data
csvpath = Path("./budget_data.csv")
#Define int to hold line number of CSV file
linenum = 0
#Define int to hold months
total_months = 0
#Define a list to hold the months
month_list = []
#Define int to hold total profit/loss
total_revenue = 0
#Define list to store CSV profit/loss data
revenue_data_list = []

#Open the CSV data as a read-only file
with open(csvpath, 'r') as csvdata:
    #Convert the data to the CSV reader
    csvreader = csv.reader(csvdata, delimiter=',')
    #Skip the first line of the CSV file
    header = next(csvreader)
    linenum += 1
    #Extract data from each row of the CSV file
    for row in csvreader:
        #Keep track of the number of months in the period
        total_months += 1
        #Find the month of each row and append them to a list
        month_list.append(str(row[0]))
        #Find the profit/loss of each row and append them to a list
        revenue_data_list.append(int(row[1]))

#Find total profit/loss over the period
for revenue in revenue_data_list:
    total_revenue += revenue

#Create variables to store the max gain/loss
max_loss = 0
max_gain = 0
#Create list to store all differences between consecutive months
diff_revenue = []
#Create dictionary to store months and gain/loss data 
diff_dict= {}

#Iterate over each month in the list in order to determine the differences
for month in range(len(month_list)-1):
    #Find the difference between two consecutive months
    difference = revenue_data_list[month+1] - revenue_data_list[month]
    #Append the differences to a list
    diff_revenue.append(difference)
    #Use the month and differences lists to fill the dictionary
    diff_dict [month_list[month]] = diff_revenue[month]
    #Determine the maximum loss
    if (difference < max_loss):
        max_loss = difference
    #Determine the maximum gain
    elif difference > max_gain:
        max_gain = difference

#Create variable to store the sum of differences
total_diff = 0
#Determine the average difference between two months
for diff in diff_revenue:
    total_diff += diff
#Calculate the average, with one less diff than the total months
average_diff = round(total_diff/(total_months-1), 2)
      
#Find the month associated with max gains/losses
def get_key(max_revenue):
    for month, revenue in diff_dict.items():
        if max_revenue == revenue:
            return month
    return "Key doesn't exist."

#Print statements for terminal
print("Finacial Analysis")
print("--------------------------")     
print(f"Total Months: {total_months}")
print(f"Total: ${total_revenue}")
print (f"Average change: ${average_diff}")
print(f"Greatest Increase in Profits: {get_key(max_gain)} (${max_gain})")
print(f"Greatest Decrease in Profits: {get_key(max_loss)} (${max_loss})")

#Create text to be printed to output file
text = ["Finacial Analysis", "--------------------------", f"Total Months: {total_months}", f"Total: ${total_revenue}", f"Average change: ${average_diff}", 
        f"Greatest Increase in Profits: {get_key(max_gain)} (${max_gain})", f"Greatest Decrease in Profits: {get_key(max_loss)} (${max_loss})"]

# Set the output file path
output_path = Path('./output.txt')
#Print text to output file
with open(output_path, 'w') as output:
    output.writelines('\n'.join(text))
```
